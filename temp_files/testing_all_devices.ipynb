```python
# Подключение мощемера (pm400)
import time
import numpy as np
pm_device=PMDevicePM100D()
def measure_average_power(pm_device):
    duration = 1
    aver_point = 3
    power_values = []
    i = 0
    for i in range(aver_point):
        power = pm_device.get_power()
        power_values.append(power)
        time.sleep(duration/aver_point)
    avg_pm_power=np.mean(power_values) * 1000 # сразу переводим на мВт
    return float(avg_pm_power)
pm_power = measure_average_power(pm_device)

print('current pm power =',pm_power,'mW')
```
