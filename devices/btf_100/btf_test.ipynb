```python
# Ячейка 1: Импорт и создание экземпляра класса
from btf_100 import BTF100
import time

# Укажите ваш COM-порт (например, 'COM6' для Windows или '/dev/ttyUSB0' для Linux)
COM_PORT = 'COM12'  # ⚠️ ИЗМЕНИТЕ НА ВАШ ПОРТ

# Создаем экземпляр класса
btf = BTF100(port=COM_PORT)
print("✅ Объект BTF100 создан")
```

```python
# Ячейка 2: Тест подключения (функция connect)
print("🔄 Попытка подключения...")
result = btf.connect()

if result:
    print("✅ connect() - УСПЕШНО")
else:
    print("❌ connect() - ОШИБКА")
```

```python
# Ячейка 3: Тест get_info()
if btf._connected:
    print("🔄 Получение информации об устройстве...")
    info = btf.get_info()
    print(f"✅ get_info():\n{info}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 4: Тест get_wavelength()
if btf._connected:
    print("🔄 Получение текущих параметров...")
    wl, bw = btf.get_wavelength()
    if wl is not None:
        print(f"✅ get_wavelength(): WL={wl:.2f} нм, BW={bw:.2f} нм")
    else:
        print("❌ get_wavelength() - ОШИБКА")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
#Тест set_wavelength() - только длина волны
if btf._connected:
    test_wl = 1550.0  # Тестовая длина волны
    print(f"🔄 Установка длины волны {test_wl} нм...")
    try:
        btf.set_wavelength(test_wl)
        time.sleep(1)  # Даем время на установку
        wl, bw = btf.get_wavelength()
        print(f"✅ Текущие параметры: WL={wl:.2f} нм, BW={bw:.2f} нм")
    except Exception as e:
        print(f"❌ set_wavelength() - ОШИБКА: {e}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Тест set_linewidth() - только ширина полосы
if btf._connected:
    test_bw = 2.0  # Тестовая ширина полосы
    print(f"🔄 Установка ширины полосы {test_bw} нм...")
    try:
        btf.set_linewidth(test_bw)
        time.sleep(1)
        wl, bw = btf.get_wavelength()
        print(f"✅ Текущие параметры: WL={wl:.2f} нм, BW={bw:.2f} нм")
    except Exception as e:
        print(f"❌ set_linewidth() - ОШИБКА: {e}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 7: Тест set_wavelength_and_linewidth() - оба параметра
if btf._connected:
    test_wl = 1555.0
    test_bw = 8.5
    print(f"🔄 Установка WL={test_wl} нм, BW={test_bw} нм...")
    try:
        btf.set_wavelength_and_linewidth(test_wl, test_bw)
        time.sleep(1)
        wl, bw = btf.get_wavelength()
        print(f"✅ Текущие параметры: WL={wl:.2f} нм, BW={bw:.2f} нм")
    except Exception as e:
        print(f"❌ set_wavelength_and_linewidth() - ОШИБКА: {e}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 8: Тест с граничными значениями
if btf._connected:
    print("🔄 Тестирование граничных значений...")
    
    # Минимальные значения
    try:
        btf.set_wavelength_and_linewidth(1525.0, 1.0)
        time.sleep(1)
        wl, bw = btf.get_wavelength()
        print(f"✅ MIN: WL={wl:.2f} нм, BW={bw:.2f} нм")
    except Exception as e:
        print(f"❌ MIN: {e}")
    
    # Максимальные значения
    try:
        btf.set_wavelength_and_linewidth(1565.0, 18.0)
        time.sleep(1)
        wl, bw = btf.get_wavelength()
        print(f"✅ MAX: WL={wl:.2f} нм, BW={bw:.2f} нм")
    except Exception as e:
        print(f"❌ MAX: {e}")
    
    # Возврат к средним значениям
    btf.set_wavelength_and_linewidth(1550.0, 5.0)
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 9: Тест обработки ошибок (некорректные значения)
if btf._connected:
    print("🔄 Тестирование обработки ошибок...")
    
    # Некорректная длина волны
    try:
        btf.set_wavelength(1500.0)  # Меньше минимума
        print("❌ Должна была возникнуть ошибка!")
    except ValueError as e:
        print(f"✅ Корректная ошибка: {e}")
    
    # Некорректная ширина полосы
    try:
        btf.set_linewidth(20.0)  # Больше максимума
        print("❌ Должна была возникнуть ошибка!")
    except ValueError as e:
        print(f"✅ Корректная ошибка: {e}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 10: Тест многократного чтения параметров
if btf._connected:
    print("🔄 Многократное чтение параметров...")
    for i in range(5):
        wl, bw = btf.get_wavelength()
        print(f"  Попытка {i+1}: WL={wl:.2f} нм, BW={bw:.2f} нм")
        time.sleep(0.5)
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 11: Тест отключения (disconnect)
if btf._connected:
    print("🔄 Отключение...")
    btf.disconnect()
    print("✅ disconnect() выполнен")
    
    # Проверка, что команды больше не работают
    try:
        btf.get_info()
        print("❌ Должна была возникнуть ошибка!")
    except RuntimeError as e:
        print(f"✅ Корректная ошибка после отключения: {e}")
else:
    print("⚠️ Сначала выполните connect()")
```

```python
# Ячейка 12: Полный цикл подключение-работа-отключение
print("🔄 Полный цикл работы:")

# Подключение
print("1. Подключение...")
btf.connect()

if btf._connected:
    # Получение информации
    print("2. Информация об устройстве:")
    print(f"   {btf.get_info()}")
    
    # Получение текущих параметров
    wl, bw = btf.get_wavelength()
    print(f"3. Текущие параметры: {wl} нм, {bw} нм")
    
    # Установка новых параметров
    print("4. Установка новых параметров...")
    btf.set_wavelength_and_linewidth(1540.0, 7.5)
    
    # Проверка установки
    wl, bw = btf.get_wavelength()
    print(f"5. Новые параметры: {wl} нм, {bw} нм")
    
    # Отключение
    print("6. Отключение...")
    btf.disconnect()
    
    print("✅ Полный цикл завершен")
else:
    print("❌ Не удалось подключиться")
```

```python
# Ячейка 1: Поиск доступных COM-портов
import serial.tools.list_ports

print("🔍 Поиск доступных COM-портов...")
ports = serial.tools.list_ports.comports()

if not ports:
    print("❌ Нет доступных COM-портов")
else:
    print(f"✅ Найдено {len(ports)} портов:")
    for i, port in enumerate(ports):
        print(f"  {i+1}. {port.device} - {port.description}")
        # Проверяем, не занят ли порт
        try:
            test_ser = serial.Serial(port.device, timeout=1)
            test_ser.close()
            print(f"     ✓ Порт {port.device} свободен")
        except serial.SerialException as e:
            print(f"     ✗ Порт {port.device} ЗАНЯТ: {e}")
```
