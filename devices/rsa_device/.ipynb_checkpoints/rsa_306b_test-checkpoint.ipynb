```python
#Подключение РЧ анализатора
from RF306B import RF306B 
rf_device=RF306B()
```

```python
# Проверка установки центральной частоты (Center Frequency)
new_cf = 2.5e9
rf_device.set_cf(new_cf)
print(f"Center Frequency установлен в: {rf_device.cf} Гц")
```

```python
# Проверка уровня опорного сигнала
new_ref = -10
rf_device.set_refLevel(new_ref)
print(f"Ref Level установлен в: {rf_device.refLevel} дБм")
```

```python
# Проверка Span (диапазон развертки)
new_span = 10e6
rf_device.set_span(new_span)
print(f"Span установлен в: {rf_device.span} Гц")
```

```python
# Проверка RBW (полоса разрешения)
new_rbw = 100e3
rf_device.set_rbw(new_rbw)
print(f"RBW установлен в: {rf_device.rbw} Гц")
```

```python
# Проверка Start/Stop частот
new_start = 2.4e9
new_stop = 2.6e9
rf_device.set_start(new_start)
rf_device.set_stop(new_stop)
print(f"Start: {rf_device.start}, Stop: {rf_device.stop}")
```

```python
try:
    freq, trace = rf_device.get_rf()
    
    # Проверка типов данных
    print(f"Тип freq: {type(freq)}, Размер: {freq.shape}")
    print(f"Тип trace: {type(trace)}, Размер: {trace.shape}")
    
    # Построение графика
    plt.figure(figsize=(12, 6))
    plt.plot(freq, trace, label='Spectrum Trace')
    plt.title(f'Spectrum Analyzer Trace (CF: {rf_device.cf/1e9} GHz)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude (dBm)')
    plt.grid(True)
    plt.legend()
    plt.show()
    
except Exception as e:
    print(f"Ошибка при получении данных: {e}")
```

```python
# Явное удаление объекта для вызова __del__ и отключения прибора
del rf_device
print("Объект удален, соединение должно быть разорвано.")
```
