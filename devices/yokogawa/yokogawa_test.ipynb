```python
import numpy as np
import matplotlib.pyplot as plt
from OSA_Yokogawa_new import OSA_Yokogawa_new

# Создание экземпляра класса (автоматически устанавливает соединение)
osa = OSA_Yokogawa_new()

# Проверка подключения (вывод идентификации устройства)
osa.idn()
```

```python
# Установка диапазона длин волн
osa.set_start('1500nm')      # Начальная длина волны
osa.set_stop('1600nm')       # Конечная длина волны
osa.set_center('1550nm')     # Центральная длина волны
osa.set_span('100nm')        # Диапазон измерения

# Установка разрешения
osa.set_resolution('0.100nm')  # Разрешение по длине волны
```

```python
# Выбор масштаба по оси Y
osa.set_log_scale()   # Логарифмический масштаб (dB)
# или
osa.set_lin_scale()   # Линейный масштаб
```

```python
# Режим чувствительности
osa.set_sens_mode('normal')  # Варианты: 'nhld', 'naut', 'normal', 'mid'

# Скорость сканирования
osa.set_sweep_speed('2x')    # Варианты: '1x', '2x'
```

```python
# Выполнение калибровки нуля (блокирует устройство на время процедуры)
osa.peform_zeroing()

# ИЛИ отключение автоматической калибровки (каждые ~5 мин)
osa.disable_zeroing()
```

```python
# Запуск измерения и получение данных X (длина волны) и Y (мощность)
x, y = osa.get_os()

# Отображение графика
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.xlabel('Длина волны (nm)')
plt.ylabel('Мощность (dBm)')
plt.title('Спектральный анализ')
plt.grid(True)
plt.show()
```

```python
# Получение данных без запуска нового измерения (использует последние данные)
x, y = osa.get_data()

# Или только Y-данные
y_only = osa.get_os_y()
```

```python
# Закрытие соединения (вызывается автоматически при удалении объекта)
del osa

# Или явно закрыть устройство
osa.device.close()
```
