```python
# ОБЩИЕ КОМАНДЫ
COMMAND = "?"           # Список всех поддерживаемых команд
COMMAND = "RESET"       # Программный сброс прибора
COMMAND = "Q?"          # Запрос текущего статуса прибора (IDLE/BUSY/ERROR)
COMMAND = "SLEEP"       # Перевод устройства в спящий режим
```

```python
#  ВЕРСИЯ И ИНФОРМАЦИЯ
COMMAND = "V1"          # Запрос версии прошивки (Firmware Version)
COMMAND = "V2"          # Запрос серийного номера устройства
COMMAND = "D?"          # Запрос даты производства (DD/MM/YY)
```

```python
# НАСТРОЙКИ RS-232 (СКОРОСТЬ)
COMMAND = "B1"          # Установка скорости 9600 бод (по умолчанию)
COMMAND = "B2"          # Установка скорости 19200 бод
COMMAND = "B3"          # Установка скорости 38400 бод
COMMAND = "B4"          # Установка скорости 57600 бод
COMMAND = "B5"          # Установка скорости 115200 бод
```

```python
# I²C АДРЕС
COMMAND = "A?"          # Запрос текущего I²C адреса
COMMAND = "A<nn>"       # Установка I²C адреса (nn = 00-127)
```

```python
# СКОРОСТЬ МОТОРА
COMMAND = "K?"          # Запрос текущей скорости мотора
COMMAND = "K<nnnn>"     # Установка скорости мотора (nnnn = значение)
```

```python
#  ПОЗИЦИЯ В ШАГАХ
COMMAND = "S?"          # Запрос текущей позиции в шагах
COMMAND = "S<nnnnnn>"   # Установка абсолютной позиции в шагах
COMMAND = "S+<nnnnnn>"  # Увеличить позицию на N шагов
COMMAND = "S-<nnnnnn>"  # Уменьшить позицию на N шагов
```

```python
# ВРЕМЯ ЗАДЕРЖКИ (ОСНОВНЫЕ КОМАНДЫ)
COMMAND = "T?"          # Запрос текущей задержки в пикосекундах
COMMAND = "T<tttt.ttt>" # Установка абсолютной задержки (например T100.000)
COMMAND = "T0"          # Установить минимальную задержку
COMMAND = "T+<tttt.ttt>"# Увеличить задержку на значение (например T+10.000)
COMMAND = "T-<tttt.ttt>"# Уменьшить задержку на значение (например T-10.000)
```

```python
# МИНИМАЛЬНАЯ ЗАДЕРЖКА
COMMAND = "M?"          # Запрос минимальной задержки
COMMAND = "M<tttt.ttt>" # Установка минимальной задержки
```

```python
# ИНТЕРФЕЙС И НАПРЯЖЕНИЕ
COMMAND = "I?"          # Запрос статуса напряжения интерфейса
COMMAND = "I0"          # Установить TTL уровень (5V)
COMMAND = "I1"          # Установить RS-232 уровень (3.3V)
```

```python
# СТАТУС ШИНЫ ИНТЕРФЕЙСА
COMMAND = "J?"          # Запрос статуса шины интерфейса
COMMAND = "J0"          # Выключить шину
COMMAND = "J1"          # Режим SPI
COMMAND = "J2"          # Режим I²C
```

```python
import serial
import time
TIMEOUT = 3
COM_PORT = "COM10"
BAUDRATE = 9600
ser = serial.Serial(port=COM_PORT, baudrate=BAUDRATE, bytesize=serial.EIGHTBITS, parity=serial.PARITY_NONE, stopbits=serial.STOPBITS_ONE, timeout=TIMEOUT, xonxoff=False, rtscts=False, dsrdtr=False)
print(f"Порт {COM_PORT} открыт\n")
```

```python
TIMEOUT = 2

# COMMAND = "gf"
COMMAND = "gr"
# Отправка команды
ser.write(f"{COMMAND}\r\n".encode('ascii'))

print(f"Отправлено: {COMMAND}")
# Чтение ответа
response = ""
start = time.time()

while time.time() - start < TIMEOUT:
    if ser.in_waiting > 0:
        line = ser.readline().decode('ascii', errors='ignore').strip()
        if line:
            print(f"Получено: {line}")
            response += line + "\n"

if response:
    print("ПРИБОР ОТВЕТИЛ")
else:
    print("ПРИБОР НЕ ОТВЕТИЛ")
```
